```python
#import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#matplotlib and seaborn for any visualisations required
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
DataSet = make_blobs(n_samples=300, n_features=3, 
                           centers=4, cluster_std=1.8,random_state=101)
#this creates a data set of 200 samples and 3 features
```

```python
DataSet
```

```python
#DataSet[1] takes 4 different values
plt.scatter(DataSet[0][:,0],DataSet[0][:,1],c=DataSet[1],cmap='rainbow')
#visualise the cluster representation of DataSet[0] by sequence of colors as in DataSet[1]
```

```python
#creating the clusters 
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(DataSet[0])#fitting based on data of DataSet[0]
```

```python
kmeans.cluster_centers_#check our cluster centers
```

```python
kmeans.labels_
```

```python
#we can 
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(DataSet[0][:,0],DataSet[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(DataSet[0][:,0],DataSet[0][:,1],c=DataSet[1],cmap='rainbow')
```

```python
#we can see from these both figures that our DataSet[0] is clustered very precisely although our colour scalings might be different.
```
